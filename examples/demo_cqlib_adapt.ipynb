```python
from quantum_hw import QuantumHardwareClient
from quantum_hw import QuantumCircuit
```

```python
qc = QuantumCircuit(2)
qc.cz(0, 1)
qc.barrier()
qc.measure_all()
qc.draw()
```

```python
client = QuantumHardwareClient()

res = client.run_auto(
    circuit=qc,
    name='demo_cqlib_provider',
    num_qubits=2,
    provider='cqlib',
    cqlib_platform='tianyan',
    cqlib_submit_mode='submit_job',
    shots=1024,
    prefer_chips='tianyan176',
    observables=['Z0', 'Z1'],
    return_probabilities=True,
    print_true=True,
)

print('query_ids:', res.task_ids)
print('observable_values:', res.observable_values)
if res.probabilities:
    print('probabilities[0][:8]:', res.probabilities[0][:8])
```
